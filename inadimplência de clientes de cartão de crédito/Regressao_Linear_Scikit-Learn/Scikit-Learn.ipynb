## Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Criando dados sintéticos

```python
np.random.seed(1) #
x = np.random.uniform(low=0.0, high=10.0, size=(1000,))

slope = 0.25
intercept = -1.25
noise = np.random.normal(loc=0.0, scale=1.0, size=(1000,))

y = slope * x + intercept + noise
```

```python
plt.scatter(x,y, s=1);
```

## 1 - Regressão Linear com scikit-learn

### 1.1 Instanciando modelo

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.get_params()
```

### 1.2 Treinamento do modelo

```python
lr.fit(x.reshape(-1,1), y)
lr.intercept_
lr.coef_, lr.intercept_

```

### 1.3 Fazendo previsões

```python
y_pred = lr.predict(x.reshape(-1,1))
y_pred[:10]
```

### 1.4 Plotando dados e linha de regressão

```python
plt.scatter(x,y, s=1)
plt.plot(x, y_pred, 'r');
```
