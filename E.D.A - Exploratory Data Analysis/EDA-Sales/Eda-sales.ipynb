# Exploratory data analyst - Sales

### Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

```python
# Importando dataset

df = pd.read_csv("https://raw.githubusercontent.com/raulsiqueira96/Projeto_Portfolio_de_dados/main/sales.csv", low_memory=False)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.groupby(['bi_st', 'status'])['total'].sum()
```

```python
df['Name Prefix'].unique()
```

```python
df.query('full_name == "Harper, Alvaro"')
```

```python
df.groupby(['category'])['total'].sum().reset_index().sort_values(by=['total'], ascending = False)
```

```python
df.sample(5)
```

```python
df.order_date.nunique() 
```

```python
df.groupby('status')['order_id'].count().reset_index().sort_values(by= 'order_id', ascending=False)
```

```python
df.query('status == "canceled"')['payment_method'].value_counts()
```

```python
df.query('status == "complete"')['payment_method'].value_counts()
```

```python
df.query('status == "canceled"').\
    groupby(['category', 'sku'])['order_id'].\
    count().reset_index().\
    sort_values(by='order_id', ascending=False).\
    head(10)
```

```python
df.query('sku == "MATSAM59DB75ADB2F80"')['status'].value_counts()
```

```python
df.groupby('State')['total'].sum().reset_index().sort_values(by= 'total', ascending= False)
```

```python
df.query('State == "TX"').\
    groupby(['City'])['total'].\
    sum().reset_index().\
    sort_values(by= 'total', ascending=False)
```

```python
df.query('City == "Houston"').\
    groupby(['category'])['total'].\
    sum().reset_index().\
    sort_values(by= 'total', ascending=False)
```

```python
df.groupby('category')['Discount_Percent'].mean().\
    reset_index().\
    sort_values(by= 'Discount_Percent', ascending=False)
```
