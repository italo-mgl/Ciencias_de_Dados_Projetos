## Criando sistema de recomendação de filmes utilizando modelo KNN (K-nearest neighbors)

```python
## Importando bibliotecas

import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
## Importando dados e Filtrando apenas as colunas que serão usadas
df_movies = pd.read_csv('https://github.com/krishnaik06/Recommendation_complete_tutorial/raw/master/KNN%20Movie%20Recommendation/movies.csv')
df_movies = df_movies[['movieId','title']]
df_movies
```

```python
## Importando dados de avaliação (Rating) dos filmes
df_ratings = pd.read_csv('https://github.com/krishnaik06/Recommendation_complete_tutorial/raw/master/KNN%20Movie%20Recommendation/ratings.csv')
df_ratings
```

```python
## Filtrando apenas as colunas que serão usadas
df_ratings = df_ratings[['userId', 'movieId', 'rating']]
df_ratings
```

```python
## Fazendo o merge das tabelas
df = pd.merge(df_movies, df_ratings, on='movieId')
df
```

```python
df_recomender = df.pivot_table(index='userId', columns='title', values='rating').fillna(0)
df_recomender
```

#### Aplicando o modelo KNN

```python
modelo_knn = NearestNeighbors(metric='cosine')
modelo_knn.fit(df_recomender)
```

```python
idx_usuario = 230

distancia, indices_vizinhos = modelo_knn.kneighbors(df_recomender.iloc[idx_usuario].values.reshape(1,-1), n_neighbors=3)
```

```python
usuario = df_recomender.index[idx_usuario]
for i in range(0, len(distancia.flatten())):
    if i == 0:
        print(f"Usuário: {idx_usuario}\n")
    else:
        print(f"Vizdinho {df_recomender.index[indices_vizinhos.flatten()[i]]} com distância de {distancia.flatten()[i]}.")
```

```python
df_usuario = df_recomender.loc[usuario].to_frame()
vizinho_proximo = df_recomender.index[indices_vizinhos.flatten()[1]]
df_vizinho = df_recomender.loc[vizinho_proximo].to_frame()
```

```python
df_titulos = pd.merge(df_usuario, df_vizinho, on='title').sort_values(by=vizinho_proximo, ascending=False)
```

```python
df_titulos = df_titulos[(df_titulos[vizinho_proximo] > 0 ) & (df_titulos[usuario] == 0)].reset_index()
df_titulos
```
