## Tabela original em HTML vs DataFrame final feito com PANDAS

### Importando bibliotecas

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from io import StringIO
```

### Solicitando a URL

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2020_totals.html'
req = requests.get(url)
print(req) ## Verificando se esta tudo ok.
```

### Analisando Pagina HTML


```python
soup = BeautifulSoup(req.content, 'html.parser')
```

### Encontrando a tabela

```python
tabela = soup.find(name='table')
```

### Transformando tabela HTML para um DataFrame em Pandas

```python
df = pd.read_html(str(tabela))[0].set_index('Rk')
df.head()
## A utilizao do "[0]"  necessrio para que o resultado seja uma lista com elementos nicos, esses elementos so o que queremos, no a lista.
```
